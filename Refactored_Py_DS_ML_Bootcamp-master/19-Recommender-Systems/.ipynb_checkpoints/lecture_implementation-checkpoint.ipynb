```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
column_names =['user_id','item_id','rating','timestamp']
data = pd.read_csv('u.data',sep='\t',names= column_names)
```

```python
data.head()
```

```python
title = pd.read_csv('Movie_Id_Titles')
```

```python
title.head()
```

```python
movies = pd.merge(data,title,on='item_id')
movies.head(1)
num_of_rating = movies.groupby('title')['rating'].count();
```

```python
rating = movies.groupby('title')['rating'].mean()
ratings = pd.DataFrame()
ratings['rating'] = rating
ratings['num_of_rating'] = num_of_rating
ratings.head()
```

```python
ratings.hist('num_of_rating')
```

```python
sns.jointplot(x='rating',y='num_of_rating',data=ratings)
```

```python
ratings.head()
movies.head()
moviemat = movies.pivot_table(index='user_id',columns='title',values = 'rating')
```

```python
moviemat.head()
```

```python
ratings[ratings['num_of_rating']>100].sort_values('num_of_rating',ascending=False).head()
```

```python
starwars_rating = moviemat['Star Wars (1977)']
```

```python
starwars_rating.head()
```

```python
starwars_rating.dropna(inplace=True)
```

```python
starwars_rating
```

```python
liarliar_rating = moviemat['Liar Liar (1997)']
```

```python
liarliar_rating.dropna(inplace=True)
```

```python
similar_to_starwars = moviemat.corrwith(starwars_rating)
```

```python
similar_to_starwars
```

```python
similar_to_liarliar = moviemat.corrwith(liarliar_rating)
```

```python
similar_to_liarliar
```

```python
similar_to_starwars = pd.DataFrame(similar_to_starwars,columns=['Corrrelation'],index=similar_to_starwars.index)
```

```python
similar_to_starwars =similar_to_starwars.join(ratings['num_of_rating'])
```

```python
similar_to_starwars[similar_to_starwars['num_of_rating']>100].sort_values('Corrrelation',ascending=False)
```
