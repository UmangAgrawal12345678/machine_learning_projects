```python
fig = plt.figure();
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Head Title')
axes2.set_title('Sub plot title');
```

```python
fig= plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='X label');
axes.plot(x,x**3,label='X Cube label');
axes.legend(loc=(0.1,0.1))
```

```python
fig = plt.figure();
axis = fig.add_axes([0,0,1,1])
axis.plot(x,y,color='purple',lw=5,ls= '-',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=2,
         markeredgecolor='green')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,lw=2,ls='--')
axes.set_xlim([1,2])
axes.set_ylim([1,3])
```
