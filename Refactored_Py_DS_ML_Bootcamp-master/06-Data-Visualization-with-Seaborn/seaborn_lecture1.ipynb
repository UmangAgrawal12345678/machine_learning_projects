```python
import seaborn as sns;
import matplotlib.pyplot as plt;
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins= 30)
```

```python
sns.jointplot('total_bill','tip',tips)
```

```python
sns.pairplot(tips,hue='sex',palette='Accent_r')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
#Categoriacal plots
import numpy as np;
```

```python
sns.barplot(x='day',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='sex',y='total_bill',data=tips,hue='day')
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
#gridplot
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
gp = sns.PairGrid(iris)
gp.map(plt.scatter)
gp.map_diag(sns.distplot)
gp.map_lower(sns.swarmplot)
```

```python
fg = sns.FacetGrid(tips,col='time',row='size')
fg.map(sns.distplot,'total_bill')
```

```python
#Regression Plots
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='day',hue='smoker',aspect=0.6,height=10)
```
